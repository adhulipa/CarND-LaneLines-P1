## Helper Functions
Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math
import numpy.linalg as linalg

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    #return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # Adi's algorithm
    red_color = [255, 0, 0]
    green_color = [0, 255, 0]
    horizontal_line_slope = 0.0
    
    # Inner helper functions
    def slope(line):
        for x1,y1,x2,y2 in line:
            return (y2-y1)/(x2-x1)
    
    def draw_line(line, color, thickness):
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1,y1), (x2,y2), color, thickness)
        return

    def bifurcate_lane_lines(lines):
        """
        This function take a group of lines and separates them into the left & right lane lines.
        Returns a pair (2-tuple) of list of lines as (left_lines, right_lines).
        """
        left_lines = []
        right_lines = []
        for line in lines:
            line_slope = slope(line)
            if (math.isclose(line_slope, horizontal_line_slope)):
                continue # i.e. ignore horizontal lines
            if line_slope < 0.0:
                left_lines.append(line)
            else:
                right_lines.append(line)
        return (left_lines, right_lines)

    def combine_similar_lines(lines):
        def calc_rho(line, line_slope):
            m = abs(line_slope)
            theta = np.arctan(m)
            rho = lambda x,y: x*np.cos(theta) + y*np.sin(theta)
            for x1,y1,x2,y2 in line:
                return (rho(x1,y1) + rho(x2,y2)) / 2
        
        similar_lines = []
        for line1 in lines:
            for line2 in lines:
                if np.array_equal(line1, line2):
                    continue # skip if it's the same line
                m1 = slope(line1)
                m2 = slope(line2)
                if math.isclose(abs(m1-m2),0.0,abs_tol=0.01):
                    rho1 = calc_rho(line1, m1)
                    rho2 = calc_rho(line2, m2)
                    if math.isclose(rho1,rho2,rel_tol=0.1):
                        avg_line = (line1 + line2) / 2
                        avg_line = avg_line.astype(np.int32)
                        similar_lines.append(avg_line)
        return similar_lines
    
    def define_lane_segment():
        # TODO
        # Combine similar lines into a single line.
        # Find the point closest to x-axis and the point closes to y-axis
        closest_to_x = (float('inf'), float('inf'))
        closest_to_y = (float('inf'), float('inf'))
        for line in similar_lines:
            for x1,y1,x2,y2 in line:
                if x1 < closest_to_x[0]:
                    closest_to_x = np.array([x1,y1], dtype=np.int32)
                elif x2 < closest_to_x[0]:
                    closest_to_x = np.array([x2,y2], dtype=np.int32)

                if y1 < closest_to_y[1]:
                    closest_to_y = np.array([x1,y1], dtype=np.int32)
                elif y2 < closest_to_y[0]:
                    closest_to_y = np.array([x2,y2], dtype=np.int32)
        line_seg = np.array([[closest_to_x[0], closest_to_x[1], closest_to_y[0], closest_to_y[1]]], dtype=np.int32)

    
    # Separate lines into left & right lines    
    (left_lines, right_lines) = bifurcate_lane_lines(lines)
    
    # Combine similar line segments
    # Line segments are similiar if they have same slope and are equidistant from origin.
    # left_lines = combine_similar_lines(left_lines)
    # right_lines = combine_similar_lines(right_lines)

    for line in left_lines:
        draw_line(line, red_color, thickness)
    
    for line in right_lines:
        draw_line(line, green_color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.
def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

## Pipeline


````python
def pipeline(image):
    """
    Applies an image processing pipeline of the sequence:
      * Graysacle
     -> Canny edge detection
     -> Region selection
     -> Hough transform
    returns: a tuple of the final result along with the intermiediate result for each step in the pipeline.
    """
    # Copy image
    initial_image = np.copy(image)
    
    # Color select
    # Select only white & yellow pixels
    white_color_select = np.copy(image)
    yellow_color_select = np.copy(image)

    white_color = [200, 200, 200]
    yellow_color_high = [220, 170, 90]
    yellow_color_low = [190, 160, 35]
    
    # White color select
    white_thresholds =  (image[:,:,0] < white_color[0]) | (image[:,:,1] < white_color[1]) | (image[:,:,2] < white_color[2])
    white_color_select[white_thresholds] = [0,0,0]
    
    # Yellow select
    image = np.copy(initial_image)
    yellow_high_thresholds = (image[:,:,0] < yellow_color_high[0]) | (image[:,:,1] < yellow_color_high[1]) | (image[:,:,2] < yellow_color_high[2])
    yellow_low_thresholds = (image[:,:,0] > yellow_color_low[0]) | (image[:,:,1] > yellow_color_low[1]) | (image[:,:,2] > yellow_color_low[2])
    yellow_thresholds = yellow_low_thresholds & yellow_high_thresholds
    yellow_color_select[yellow_thresholds | ~white_thresholds] = [0,0,0]
    
    # Combine both white & yellow selections
    image = white_color_select[:,:,:] + yellow_color_select[:,:,:]
    color_image = np.copy(image)
    
    # Convert to grayscale
    image = grayscale(image)
    gray_image = np.copy(image)

    # Detect Edges
    # Apply a gaussian blur
    image = gaussian_blur(image, 3)

    # Apply edge detection
    image = canny(image,
                  low_threshold=50,
                  high_threshold=150)
    edges_image = np.copy(image)

    # Select region
    (image_height, image_width) = image.shape
    half = 1 / 2
    third = 1 / 3
    slack = 50 # px

    # Define a region-of-interest to be below the horizon
    vertices = np.array([[
        # bottom left
        (3*slack, image_height),

        # top left
        (third*image_width + slack, half*image_height + 1.5*slack),

        # top right
        (2*third*image_width + slack, half*image_height + 1.5*slack),

        # bottom right
        (image_width - slack, image_height)
    ]], dtype=np.int32)

    image = region_of_interest(image, vertices)
    image_cropped_regions = np.copy(image)

    # Run Hough transform
    image = hough_lines(image, rho=1 , theta=(np.pi / 180),
                        threshold=12, min_line_len=5, max_line_gap=8)
    hough_image = np.copy(image)

    # Overlay lines over initial image
    overlay_image = np.copy(weighted_img(image, initial_image))
    
    return overlay_image, color_image, gray_image, edges_image, image_cropped_regions, hough_image
````
